```python
inputs = []
outputs = []
for word in data["content"][0:200]:
    for i in range(len(token.texts_to_sequences([word])[0]) -1):
        inputs.append(token.texts_to_sequences([word])[0][:i+1])
        outputs.append(token.texts_to_sequences([word])[0][i+1])
```

```python
print(inputs)
print(outputs)
```

```python
from keras.utils import pad_sequences
```

```python
fv=pad_sequences(inputs)
```

```python
fv.shape
```

```python
fv
```

```python
from keras.utils import to_categorical
```

```python
cv=to_categorical(outputs,num_classes=1103)
```

```python
cv
```

```python
cv.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding,Input
```

```python
model=Sequential()
model.add(Embedding(1103,10,input_shape=(114,)))
model.add(LSTM(200,activation="tanh",return_sequences=False))
model.add(Dense(1103,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(fv,cv,epochs=100,batch_size=50,validation_split=0.2)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

text="Please"
for i in range(5):
    #tokenize the input_text
    sequences=token.texts_to_sequences([text])
    
    #pad the sequences to ensure consistant input size for the model
    padded_sequences=pad_sequences(sequences,maxlen=114)
    
    #predict the next word
    predicted_probs=model.predict(padded_sequences)
    v=np.argmax(predicted_probs)
    
    #get the word corresponding to the predicted index
    predicted_word=token.index_word.get(v,'') #use .get to handle out of vocubulary indices
    
    #append the predicted word to the text
    text=text + " " + predicted_word
    print(text)
```
